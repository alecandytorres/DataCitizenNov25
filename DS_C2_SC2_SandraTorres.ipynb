#  b) Estructura y Tipo de Datos
*   Transformación de columnas de fecha a tipo datetime64.
*   Mostrar la estructura y tipo de datos de cada columna para identificar qué operaciones puedes realizar con cada una de ellas.
*   Asegurar que las columnas con fechas sean del tipo datetime64.



```python
archivo['date'] = pd.to_datetime(archivo['date'])
archivo.info()
```

# c) Cantidad de Vacunas Aplicadas por Compañía.
*   Determinar la cantidad de vacunas aplicadas de cada compañía (con base en cómo lo reporta cada país en la columna vaccines).


*   Agrupe por vaccines y realice la sumatoria.

```python
archivo[["vaccines","daily_vaccinations"]].groupby("vaccines").sum()
```

# d) Cantidad Total de Vacunas Aplicadas
* Obtener la cantidad de vacunas aplicadas en todo el mundo.

```python
print("Vacunación Total: ", archivo["daily_vaccinations"].sum())
```

# e) Promedio de Vacunas por País.
* Calcular el promedio de vacunas aplicadas por país.

```python
archivo[["country","daily_vaccinations"]].groupby("country").mean()
```

# f) Cantidad de Vacunas Aplicadas: **29 Ene 2021**
*   Determinar la cantidad de vacunas aplicadas el día 29/01/21 en todo el mundo.



```python
import numpy as np
vacunas_290121 = archivo["daily_vaccinations"].where(archivo["date"]== "2021-01-29", 0).sum()
print("Vacunas aplicadas 29-Ene-2021: ", vacunas_290121)
```

# g) Nuevo Dataframe **conDiferencias**
*   Crear un dataframe nuevo denominado conDiferencias que contenga los datos originales y una columna derivada (diferencias) con las diferencias de aplicación entre las columnas daily_vaccionations y daily_vaccionations_raw.  




```python
conDiferencias = archivo.copy()
conDiferencias["Diferencias"] = conDiferencias['daily_vaccinations'] - conDiferencias['daily_vaccinations_raw']
conDiferencias
```

# h) Registros de Fechas
* Obtener el periodo de tiempo entre el registro con fecha más reciente y el registro con fecha más antigua.

```python
archivo['date'].max() - archivo['date'].min()
```

# i) Nuevo Dataframe **conCantidad**
* Crear un dataframe nuevo denominado conCantidad que contenga los datos originales y una columna derivada (canVac) con la cantidad de vacunas utilizadas cada día (usar la columna vaccines y separar por el carácter , ).

```python
conCantidad = archivo.copy()
conCantidad["canVac"] = archivo['vaccines'].str.split(', ').str.len()
conCantidad
```

# j) Dataframe **antes20**
* Generar un dataframe denominado antes20 con todos los registros que se hayan realizado antes del 20 de diciembre de 2020.

```python
antes20 = archivo.query("date < '2020-12-20'")
antes20
```

# k) Dataframe **Pfizer**
* Obtener un dataframe denominado pfizer con todos los registros donde se haya utilizado la vacuna Pfizer.

```python
pfizer = archivo.loc[archivo["vaccines"].str.contains("Pfizer", case=False, na=False)].copy()
pfizer
```

# l) Almacenamiento de Dataframes en Excel
* Almacenar los dataframes generados (conDiferencias, conCantidad, antes20 y pfizer) en un archivo de Excel denominado resultadosReto.xlsx, donde cada dataframe ocupe una hoja diferente. Se recomienda ver la documentación de pd.ExcelWriter.

```python
with pd.ExcelWriter('resultadosReto.xlsx') as writer:
    conDiferencias.to_excel(writer, sheet_name='conDiferencias', index=False)
    conCantidad.to_excel(writer, sheet_name='conCantidad', index=False)
    antes20.to_excel(writer, sheet_name='antes20', index=False)
    pfizer.to_excel(writer, sheet_name='pfizer', index=False)
```
